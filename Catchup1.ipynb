```python
#1

# Create the custom exception
class NegativeBalanceError(Exception):
    pass


# Create the class TradeManager
# By default the init value for the balance will be 1000

class TradeManager:
    # create the constructor
    def __init__(self, balance=1000):
        self.__balance = balance
        self.position = dict()
        self.trans = []
        self.PNLs = dict()

    # create handle_transaction
    def handle_transaction(self, t):
        self.update_balance(t)
        self.update_position(t)
        self.update_pnl(t)

    # create update_balance
    def update_balance(self, t):
        total = t["quantity"] * t["price"]
        if t["side"] == Side.BUY:
            # if total > self.__balance:
            #     raise NegativeBalanceError
            self.set_balance(self.__balance - total)
        if t["side"] == Side.SELL:
            # if t["symbol"] not in self.position.keys() or t["quantity"] > t["symbol"]:
            #     raise NegativeBalanceError
            self.set_balance(self.__balance + total)

    # create update_pnl
    def update_pnl(self, t):
        if t["symbol"] in self.PNLs.keys():
            IN_POS = 1
        else:
            IN_POS = 0

        if t["side"] == Side.BUY and IN_POS == 0:
            self.PNLs[t["symbol"]] = - t["quantity"] * t["price"]
        elif t["side"] == Side.BUY and IN_POS == 1:
            self.PNLs[t["symbol"]] -= t["quantity"] * t["price"]
        elif t["side"] == Side.SELL and IN_POS == 0:
            self.PNLs[t["symbol"]] = t["quantity"] * t["price"]
        else:
            self.PNLs[t["symbol"]] += t["quantity"] * t["price"]

    # create update_position
    def update_position(self, t):
        if t["symbol"] in self.position.keys():
            IN_POS = 1
        else:
            IN_POS = 0

        if t["side"] == Side.BUY and IN_POS == 0:
            self.position[t["symbol"]] = t["quantity"]
        elif t["side"] == Side.BUY and IN_POS == 1:
            self.position[t["symbol"]] += t["quantity"]
        elif t["side"] == Side.SELL and IN_POS == 0:
            self.position[t["symbol"]] = -t["quantity"]
        else:
            self.position[t["symbol"]] -= t["quantity"]

    # create repr_position
    def repr_position(self):
        res = "Positions:"
        for k, v in self.position.items():
            res += k + " " + str(v) + ","
        return res[:-1]

    # create repr_pnl
    def repr_pnl(self):
        res = "PNLs:"
        for k, v in self.PNLs.items():
            res += k + " " + str(v) + ","
        return res[:-1]

    # create repr_balance
    def repr_balance(self):
        return f"Balance:{self.__balance}"

    # create set_balance
    def set_balance(self, balance):
        if balance < 0:
            raise NegativeBalanceError("Negative Balance Not Accepted")
        self.__balance = balance

    # you will certainly need to overload a built-in function
    def __repr__(self):
        return self.repr_balance() + "\n" + self.repr_position() + "\n" + self.repr_pnl()
```

```python
#2
#
# Complete the 'countDuplicate' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY numbers as parameter.
#
import collections


def countDuplicate(numbers):
    # Write your code here
    d = collections.defaultdict(int)
    count = 0
    for num in numbers:
        d[num] += 1
    for k in d.keys():
        d[k] -= 1
        if d[k] != 0:
            count += 1

    return count
```

```python
#3
def isPresent(root, val):
    # write your code here
    # return 1 or 0 depending on whether the element is present in the tree or not

    if root.value == val:
        return 1
    elif val < root.value and root.left != None:
        return isPresent(root.left, val)
    elif val > root.value and root.right != None:
        return isPresent(root.right, val)
    else:
        return 0
```
