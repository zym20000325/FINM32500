```python
# 1 Delete nodes greater than X

#!/bin/python3

import math
import os
import random
import re
import sys

class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_node(self, node_data):
        node = SinglyLinkedListNode(node_data)

        if not self.head:
            self.head = node
        else:
            self.tail.next = node

        self.tail = node

def print_singly_linked_list(node, sep, fptr):
    while node:
        fptr.write(str(node.data))

        node = node.next

        if node:
            fptr.write(sep)


#
# Complete the 'removeNodes' function below.
#
# The function is expected to return an INTEGER_SINGLY_LINKED_LIST.
# The function accepts following parameters:
#  1. INTEGER_SINGLY_LINKED_LIST listHead
#  2. INTEGER x
#

#
# For your reference:
#
# SinglyLinkedListNode:
#     int data
#     SinglyLinkedListNode next
#
#

def removeNodes(listHead, x):
    # Write your code here
    
    if not listHead:
        return None
    
    while listHead and listHead.data>x:
        listHead = listHead.next
    
    if not listHead:
        return None
    
    p = listHead.next
    q = listHead
    
    while p:
        
        if p.data > x:
            p = p.next
            q.next = p
        else:
            p = p.next
            q = q.next
    
    return listHead
            
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    listHead_count = int(input().strip())

    listHead = SinglyLinkedList()

    for _ in range(listHead_count):
        listHead_item = int(input().strip())
        listHead.insert_node(listHead_item)

    x = int(input().strip())

    result = removeNodes(listHead.head, x)

    print_singly_linked_list(result, '\n', fptr)
    fptr.write('\n')

    fptr.close()
```

```python
# 2 Shortest substring containing characters

#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'shortestSubstring' function below.
#
# The function is expected to return an INTEGER.
# The function accepts STRING givenString as parameter.
#

def shortestSubstring(givenString):
    # Write your code here
    
    n = len(givenString)
    
    if n==0:
        return 0
    
    if n==1:
        return 1
    
    char_list = list(set(list(givenString)))
    print(char_list)
    
    left = 0
    right = 0
    
    window_len = []
    
    while right<len(givenString):
        
        window = givenString[left:right+1]
        
        if len(list(set(list(window))))==len(char_list):
            window_len.append(len(window))
            left = left+1
            
        else:
            right = right+1
        
    return min(window_len)
            
    
    
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    givenString = input()

    result = shortestSubstring(givenString)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 3 Efficient Janitor

#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'efficientJanitor' function below.
#
# The function is expected to return an INTEGER.
# The function accepts FLOAT_ARRAY weight as parameter.
#

def efficientJanitor(weight):
    # Write your code here
    
    n = len(weight)
    
    if n==0:
        return 0
    
    if n==1:
        return 1
    
    weight = sorted(weight)
    
    left = 0
    right = len(weight)-1
    
    bag_num = 0
    
    while left <= right:
        
        bag_num = bag_num+1
        
        if weight[left] + weight[right] <= 3.0:
            left = left+1
            
        right = right - 1
        
    return bag_num
    

    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    weight_count = int(input().strip())

    weight = []

    for _ in range(weight_count):
        weight_item = float(input().strip())
        weight.append(weight_item)

    result = efficientJanitor(weight)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 4 Find the factor

#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'pthFactor' function below.
#
# The function is expected to return a LONG_INTEGER.
# The function accepts following parameters:
#  1. LONG_INTEGER n
#  2. LONG_INTEGER p
#

import math

def pthFactor(n, p):
    # Write your code here
    
    factors = []
    
    for i in range(int(math.sqrt(n))+1):
        
        f = i+1
        if n%f == 0:
            factors.append(f)
            if f*f!=n:
                factors.append(int(n/f))
    
    factors = sorted(factors)
            
    if p>len(factors):
        return 0
    
    else:
        return factors[p-1]
            
    
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    p = int(input().strip())

    result = pthFactor(n, p)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 5 Bit logic

#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'maxXor' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER lo
#  2. INTEGER hi
#  3. INTEGER k
#

def maxXor(lo, hi, k):
    # Write your code here
    
    max_value = 0
    
    for i in range(lo,hi):
        for j in range(i,hi+1):
            
            xor = i^j
            
            if xor==k:
                return k
            elif xor<k:
                max_value = max(max_value,xor)
    
    return max_value
                
            

    
    
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    lo = int(input().strip())

    hi = int(input().strip())

    k = int(input().strip())

    result = maxXor(lo, hi, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 6 Price check

#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'priceCheck' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. STRING_ARRAY products
#  2. FLOAT_ARRAY productPrices
#  3. STRING_ARRAY productSold
#  4. FLOAT_ARRAY soldPrice
#

def priceCheck(products, productPrices, productSold, soldPrice):
    # Write your code here
    
    product_dict = {}
    
    for i in range(len(products)):
        product_dict[products[i]] = productPrices[i]
        
    error_num = 0
    
    for i in range(len(productSold)):
        product0 = productSold[i]
        price0 = soldPrice[i]
        
        if product_dict[product0]!=price0:
            error_num = error_num+1
            
    return error_num
        
        
        
    
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    products_count = int(input().strip())

    products = []

    for _ in range(products_count):
        products_item = input()
        products.append(products_item)

    productPrices_count = int(input().strip())

    productPrices = []

    for _ in range(productPrices_count):
        productPrices_item = float(input().strip())
        productPrices.append(productPrices_item)

    productSold_count = int(input().strip())

    productSold = []

    for _ in range(productSold_count):
        productSold_item = input()
        productSold.append(productSold_item)

    soldPrice_count = int(input().strip())

    soldPrice = []

    for _ in range(soldPrice_count):
        soldPrice_item = float(input().strip())
        soldPrice.append(soldPrice_item)

    result = priceCheck(products, productPrices, productSold, soldPrice)

    fptr.write(str(result) + '\n')

    fptr.close()
```
