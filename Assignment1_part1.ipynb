```python
import yaml

def mylist1(length):
    
    result = []
    
    if(length==0):
        return result
    
    if(length%2==0):
        
        n = int(length/2)
        
        for i in range(n):
            result.append(i+1)
            result.append(-1*(i+1))
        return result
    
    else:
        
        n = int((length-1)/2)
        result.append(0)
        
        for i in range(n):
            result.append(i+1)
            result.append(-1*(i+1))
        return result
            
# parameters mylist,l,r are given you as a string
# you need to return a list of integers

'''
29, 77, 47, 97, 17, 33, 2, 51, 68, 3
47
51
'''

class ElementNotFound(Exception):
    
    def __init__(self, message="element is not found"):

        self.message = message
        super().__init__(self.message)

class ElementReversed(Exception):
    
    def __init__(self, message="elements are reversed"):

        self.message = message
        super().__init__(self.message)


def mylist2(mylist,l,r):
    
    # try:
    
    mylist = list(mylist.split(", "))
    mylist = [int(i) for i in mylist]
    l = int(l)
    r = int(r)
    
    if(l not in mylist or r not in mylist):
        raise ElementNotFound
    
    l_loc = mylist.index(l)
    r_loc = mylist.index(r)
    
    if(l_loc>=r_loc):
        raise ElementReversed
    
    if(l_loc+1==r_loc):
        return mylist
    
    result = []
    for i in range(len(mylist)):
        if(i<=l_loc or i>=r_loc):
            result.append(mylist[i])
    return result
    
                

'''
{3: [4, 5, 6], 1: [1, 2, 3], 2: [7, 8, 9]}
2

key:2,value:[7, 8]
key:1,value:[1, 2]
'''

import operator

def mydict(dic,n):
    
    # delete the largest key
    
    keys = list(dic.keys())
    max_key = max(keys)
    del dic[max_key]
    
    # delete the nth element in each list
    
    for key,value in dic.items():

        del value[n]
        dic[key] = value
    
    # display by sum in descending order
    
    sum_dic = {}
    
    for key,value in dic.items():
        
        sum0 = sum(value)
        sum_dic[key] = sum0
    
    sum_dic = dict(sorted(sum_dic.items(), key=operator.itemgetter(1),reverse=True))
    
    # print your result
    
    for key,value in sum_dic.items():
        k = key
        v = dic[k]
        print("key:"+str(k)+",value:"+str(v))
        
# parameter listsofint is a string of comma separated lists of ints, 
# two lists are separated by a semicolon
# example: "1,2,3;3,4,5;3,4,7"
# return the sorted intersection as a list
def findintersection(listsofint):
    
    lists = list(listsofint.split(";"))
    
    for i in range(len(lists)):
        lists[i] = list(lists[i].split(","))
        lists[i] = [int(i) for i in lists[i]]
    
    intersection = lists[0]
    
    for i in range(len(lists)):
        if(i>=1):
            set1 = set(intersection)
            set2 = set(lists[i])
            intersection = list(set1.intersection(set2))
    
    intersection = sorted(intersection)
    
    return intersection
    

def case1(n):
    result = mylist1(int(n))
    print(isinstance(result,list))
    print(len(result)==int(n))
    print(len(set(result))==int(n))
    print(sum(result)==0)
def case2(mylist,l,r):
    try:
        print(mylist2(mylist,l,r))
        print("Complete")
    except ElementNotFound as e:
        print(e)
    except ElementReversed as e:
        print(e)
    except Exception:
        print("Read the testcase!")
def case3(dic,n):
    dic = yaml.load(dic,Loader=yaml.FullLoader)
    mydict(dic,int(n))
def case4(listsofint):
    res = findintersection(listsofint)
    print(res)
if __name__ == "__main__":
    test_number = int(input().strip())
    num_param = int(input().strip())
    if num_param==1:
        globals()['case'+str(test_number)](input().strip())
    elif num_param==2:
        globals()['case'+str(test_number)](input().strip(),input().strip())
    elif num_param==3:
        globals()['case'+str(test_number)](input().strip(),input().strip(),input().strip())
        
```
