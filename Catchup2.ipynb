```python
#Q2 Powers of 2
def isPower(arr):
    # Write your code here
    arr = list(map(lambda x: 3 if x == 0 else x, arr))
    return list(map(lambda x: 1 if int(math.log(x,2))-math.log(x,2)==0 else 0, arr))
```

```python
from datetime import datetime

# references: homework1 part1 & homework3

# Enter your code here. Read input from STDIN. Print output to STDOUT
class NegativeNumberException(Exception):
    
    def __init__(self):
        super().__init__()


def negative_value_checker(func):
    
    def wrapper(self, *args):
       
        for a in args:
            if type(a)==float or type(a)==int:
                
                if a<0:
                    raise NegativeNumberException()
                
        return func(self,*args)
    
    return wrapper


class FuturesProduct():
    
    def __init__(self,root_symbol,tick_size,contract_value):
        
        self.root_symbol = root_symbol
        self.__tick_size = tick_size
        self.__contract_value = contract_value
        
        if self.__tick_size<0:
            raise NegativeNumberException()
        if self.__contract_value<0:
            raise NegativeNumberException()
        
    @property
    def tick_size(self):
        return self.__tick_size
    
    @property
    def contract_value(self):
        return self.__contract_value
    
    @tick_size.setter
    def tick_size(self, new_tick_size):
        
        if new_tick_size>=0:
            self.__tick_size = new_tick_size
        else:
            raise NegativeNumberException()
        
    @contract_value.setter
    def contract_value(self, new_contract_value):
        
        if new_contract_value>=0:
            self.__contract_value = new_contract_value
        else:
            raise NegativeNumberException()
    
        
class FuturesContract(FuturesProduct):
    
    def __init__(self,root_symbol,tick_size,contract_value,contract,expiration_date):
        
        super(FuturesContract, self).__init__(root_symbol,tick_size,contract_value)
        self.contract = contract
        self.expiration_date = expiration_date
    

    
    
def test_futures_product_valid_getters():
    fp = FuturesProduct("ES", 0.25, 12.5)
    assert(fp.root_symbol == "ES")
    assert(fp.tick_size == 0.25)
    assert(fp.contract_value == 12.5)

def test_futures_product_valid_setters():
    fp = FuturesProduct("", 0, 0)

    fp.root_symbol = "ES"
    fp.tick_size = 0.25
    fp.contract_value = 12.5

    assert(fp.root_symbol == "ES")
    assert(fp.tick_size == 0.25)
    assert(fp.contract_value == 12.5)

def test_futures_product_invalid_tick_size_init():
    try:
        fp = FuturesProduct("ES", -0.25, 12.5)
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_product_invalid_contract_value_init():
    try:
        fp = FuturesProduct("ES", 0.25, -12.5)
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_product_invalid_tick_size_setter():
    fp = FuturesProduct("ES", 0.25, 12.5)
    try:
        fp.tick_size = -0.25
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_product_invalid_contract_value_setter():
    fp = FuturesProduct("ES", 0.25, 12.5)
    try:
        fp.contract_value = -12.5
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_contract_valid_getters():
    fc = FuturesContract("ES", 0.25, 12.5, "ESZ0", datetime(2020, 12, 14))

    assert(fc.root_symbol == "ES")
    assert(fc.tick_size == 0.25)
    assert(fc.contract_value == 12.5)
    assert(fc.contract == "ESZ0")
    assert(fc.expiration_date == datetime(2020, 12, 14))

def test_futures_contract_valid_setters():
    fc = FuturesContract("", 0, 0, "", datetime(1970, 1, 1))

    fc.root_symbol = "ES"
    fc.tick_size = 0.25
    fc.contract_value = 12.5
    fc.contract = "ESZ0"
    fc.expiration_date = datetime(2020, 12, 14)

    assert(fc.root_symbol == "ES")
    assert(fc.tick_size == 0.25)
    assert(fc.contract_value == 12.5)
    assert(fc.contract == "ESZ0")
    assert(fc.expiration_date == datetime(2020, 12, 14))

def test_futures_contract_invalid_tick_size_init():
    try:
        fc = FuturesContract("ES", -0.25, 12.5, "ESZ0", datetime(2020, 12, 14))
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_contract_invalid_contract_value_init():
    try:
        fc = FuturesContract("ES", 0.25, -12.5, "ESZ0", datetime(2020, 12, 14))
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_contract_invalid_tick_size_setter():
    fc = FuturesContract("ES", 0.25, 12.5, "ESZ0", datetime(2020, 12, 14))
    try:
        fc.tick_size = -0.25
    except NegativeNumberException:
        return
    
    assert(False)

def test_futures_contract_invalid_contract_value_setter():
    fc = FuturesContract("ES", 0.25, 12.5, "ESZ0", datetime(2020, 12, 14))
    try:
        fc.contract_value = -12.5
    except NegativeNumberException:
        return
    
    assert(False)

@negative_value_checker
def valid_four_values(arg1, arg2, arg3, arg4):
    return True

@negative_value_checker
def invalid_four_values(arg1, arg2, arg3, arg4):
    return False

@negative_value_checker
def valid_two_values(arg1, arg2):
    return True

@negative_value_checker
def invalid_two_values(arg1, arg2):
    return False

def test_valid_decorators():
    assert(valid_four_values(None, "", 0.25, 12.5))
    assert(valid_two_values(None, 5))

def test_invalid_decorators():
    try:
        assert(invalid_four_values(None, "", -0.25, 12.5))
    except NegativeNumberException:
        pass

    try:
        assert(invalid_four_values(None, "", 0.25, -12.5))
    except NegativeNumberException:
        pass

    try:
        assert(invalid_two_values(None, -5))
    except NegativeNumberException:
        pass
    
if __name__ == "__main__":
    import sys
    func_name = sys.stdin.readline().strip()
    test_func = globals()[func_name]
    test_func()
```
