```python
import math
import os
import random
import re
import sys
import pandas as pd
import numpy as np

def case1(financial_data):
    # Print First 5 rows of MSFT
    # Print Last 5 rows of MSFT
    print(financial_data[0:5])
    print(financial_data[-5:])

def case2(financial_data):
    #Resample to monthly data mean
    #Display the first 5 rows
    print(financial_data.resample('M').mean()[0:5])

def case3(financial_data):
    # Create a variable daily_close and copy Adj Close from financial_data
    # Print first 20 daily returns
    daily_close = pd.DataFrame()
    daily_close['Adj Close'] = financial_data['Adj Close']
    # daily_close = financial_data[['Adj Close']].copy()
    daily_returns = pd.DataFrame()
    daily_returns['Adj Close'] = (daily_close['Adj Close'] - daily_close['Adj Close'].shift(1))/daily_close['Adj Close'].shift(1)
    print(daily_returns[0:20])

def case4(financial_data):
    # Calculate the cumulative daily returns
    # day1 : return1  cumulative reuturn : (1+return1)-1
    # day2 : return2  cumulative reuturn : (1+return1)*(1+return2)-1
    # Print first 20 rows
    daily_returns = pd.DataFrame()
    daily_returns['Adj Close'] = (financial_data['Adj Close'] - financial_data['Adj Close'].shift(1))/financial_data['Adj Close'].shift(1)

    cumulative_daily_returns = pd.DataFrame()
    cumulative_daily_returns['Adj Close'] = (1+daily_returns['Adj Close']).cumprod()-1
    
    print(cumulative_daily_returns[0:20])


def case5(financial_data):
    # Isolate the adjusted closing prices and store it in a variable
    # Calculate the moving average for a window of 20
    # Display the last 20 moving average number
    daily_close = pd.DataFrame()
    daily_close['Adj Close'] = financial_data['Adj Close']
    
    moving_average = pd.DataFrame()
    moving_average['Adj Close'] = daily_close['Adj Close'].rolling(20).mean()
    
    print(moving_average[-20:])
    


def case6(financial_data):
    # Calculate the volatility for a period of 100 don't forget to multiply by square root
    # don't forget that you need to use pct_change
    # Print first 20 rows (It should be last 29 rows here -Yumeng)
    volatility = pd.DataFrame()
    volatility['Adj Close'] = financial_data['Adj Close']
    volatility = volatility.pct_change().rolling(100).std()*(100**0.5)
    
    print(volatility[-20:])

    

if __name__ == '__main__':
    case_number=input().strip()
    df = pd.read_csv(sys.stdin, header=0, index_col='Date', parse_dates=True)
    globals()['case'+case_number](df)
    
```
